```r
# factors take vector of objects and returns factor
gender <- c("M", "F", "F", "O", "F", "M")
gender
```

```r
gender_factor <- factor(gender)
gender_factor
```

```r
str(gender_factor)
```

```r
# change nominal value to orginal variable by saying factor is now ordered
gender_factor_ordered <- factor(gender, ordered = TRUE)
str(gender_factor_ordered) 
```

```r
gender_factor_reordered <- factor(gender, ordered = TRUE, levels = c("O", "M", "F"))
str(gender_factor_reordered)
```

```r
# print displays the results, may or may not pass result to a variable during assignment

test_func_v1 <- function(x){
    
    # computing sum and printing
    total_val <- sum(x)
    print(total_val)
    
    # first element
    first_elem <- x[1]
    print(first_elem)
    
    first_elem_by_five <- x[1]/5
    print(first_elem_by_five)
    
    c(total_val, first_elem, first_elem_by_five)
}
```

```r
test_func_v1(c(1,2,3))
```
